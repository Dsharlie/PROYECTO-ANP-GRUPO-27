```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

variables = [ ('P1', 'Departamento' ), 
              ('P22', 'Cul fue la PRINCIPAL razn por la que empez a vivir en la calle?' ),  
              ('P28R', 'Cul es el nivel educativo MS ALTO alcanzado por usted?'), 
              ('P34', 'Usted se reconoce en su orientacin sexual como:')]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
axes = axes.flatten()

for i, (var, titulo) in enumerate(variables):
    ax = sns.countplot(x=var, data=df, ax=axes[i])
    axes[i].set_title(titulo)
    axes[i].set_xlabel('Categoras')
    axes[i].set_ylabel('Frecuencia')
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=90)
    
    # Calcular el total de observaciones para el porcentaje
    total = len(df[var])
    
    # Agregar etiquetas de porcentaje encima de las barras
    for p in ax.patches:
        percentage = 100 * p.get_height() / total  # Calcular porcentaje
        ax.annotate(f'{percentage:.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='baseline', fontsize=11, color='black', xytext=(0, 5),
                    textcoords='offset points')

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Variables y sus ttulos
variables = [ ('P9', 'Usted es hombre o mujer?'), 
              ('P13', 'Usted duerme habitualmente:')]

# Crear una figura con subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(6, 10))
axes = axes.flatten()

# Iterar sobre las variables para crear grficos de torta
for i, (var, titulo) in enumerate(variables):
    # Contar las frecuencias de cada categora
    counts = df[var].value_counts()
    labels = counts.index
    sizes = counts.values
    
    # Crear grfico de torta con colores ms fuertes
    axes[i].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, 
                colors=sns.color_palette("Set2"))
    axes[i].set_title(titulo)

# Ajustar el diseo para evitar sobreposiciones
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Asegurarse de que 'P8R' no tenga valores nulos y convertir los valores a enteros
df = df.dropna(subset=['P8R'])
df['P8R'] = df['P8R'].astype(int)

# Crear los rangos de edad en intervalos de 10 aos (15-25, 25-35, ..., 65-75)
bins = [15, 25, 35, 45, 55, 65, 75]
labels = ['15-24', '25-34', '35-44', '45-54', '55-64', '65-74']

# Asignar los rangos de edad a la columna 'Edad_Categorias'
df['Edad_Categorias'] = pd.cut(df['P8R'], bins=bins, labels=labels, right=False)

# Contar las frecuencias de cada categora
counts = df['Edad_Categorias'].value_counts().sort_index()

# Crear el grfico de barras para los rangos de edad
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=counts.index.astype(str), y=counts.values, palette='muted')

# Configurar ttulos y etiquetas
plt.title('Distribucin de Edades por Rangos de 10 aos')
plt.xlabel('Rango de Edad')
plt.ylabel('Frecuencia')

# Mostrar etiquetas de porcentaje en cada barra
total = counts.sum()
for p in ax.patches:
    percentage = 100 * p.get_height() / total
    ax.annotate(f'{percentage:.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', fontsize=11, color='black', xytext=(0, 5),
                textcoords='offset points')

# Mostrar el grfico
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
