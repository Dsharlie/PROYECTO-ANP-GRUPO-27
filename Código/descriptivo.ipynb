```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un DataFrame booleano donde True significa que el valor es nulo
plt.figure(figsize=(20, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', xticklabels=True)

plt.title('Mapa de Calor de Valores Nulos')
plt.show()
```

```python
correlation_matrix = df.corr()
correlation_matrix
```

```python
plt.figure(figsize=(25, 10))

# Crear el heatmap
sns.heatmap(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1, center=0, xticklabels=True, yticklabels=True)

# AÃ±adir el tÃ­tulo
plt.title('Mapa de Calor de CorrelaciÃ³n entre Variables')

# Mostrar el grÃ¡fico
plt.show()
```

```python
variables = [ ('P1', 'Departamento' ), ('P9', 'Â¿Usted es hombre o mujer?'), ('P13', 'Â¿Usted duerme habitualmente:'), ('P22', 'Â¿CuÃ¡l fue la PRINCIPAL razÃ³n por la que empezÃ³ a vivir en la calle?' ), 
             ('P28R', 'Â¿CuÃ¡l es el nivel educativo MÃS ALTO alcanzado por usted y el ÃšLTIMO AÃ‘O O GRADO aprobado en ese nivel?'), ('P34', 'Â¿Usted se reconoce en su orientaciÃ³n sexual como:')]

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 15))

axes = axes.flatten()

for i, (var, titulo) in enumerate(variables):
    sns.countplot(x=var, data=df, ax=axes[i])
    axes[i].set_title(titulo)
    axes[i].set_xlabel('CategorÃ­as')
    axes[i].set_ylabel('Frecuencia')
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=90)

plt.tight_layout()
plt.show()
```

```python
columnas = ['P1', 'P9', 'P13', 'P22', 'P28R', 'P34' ]

# Contar las categorÃ­as en cada columna
for col in columnas:
    print(f'CategorÃ­as en {col}:')
    print(df[col].value_counts())
    print('\n')
```

```python
for col in columnas:
    print(f'Porcentaje de categorÃ­as en {col}:')
    print(df[col].value_counts(normalize=True) * 100)
    print('\n')
```

```python
df_melt = df[['P26S1', 'P26S2', 'P26S3', 'P26S4', 'P26S5', 'P26S6']].melt(var_name='Variable', value_name='CategorÃ­a')

# Crear el grÃ¡fico de barras apilado
df_melt['Variable'] = df_melt['Variable'].replace({
    'P26S1': 'Familiar',
    'P26S2': 'Amigo',
    'P26S3': 'Instituciones Oficiales',
    'P26S4': 'InstituciÃ³n Privada',
    'P26S5': 'OrganizaciÃ³n Religiosa',
    'P26S6': 'Otros'
})

sns.countplot(data=df_melt, x='CategorÃ­a', hue='Variable')

plt.title('Â¿Usted recibe algÃºn tipo de ayuda:')
plt.xlabel('CategorÃ­a')
plt.ylabel('Frecuencia')
plt.legend(title='Variable')
plt.show()
```
